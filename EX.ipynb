```python
## TUPLE
# 建立
a = 1, #方法1
b = (1) #方法2 
```

```python
## LIST
# 建立
a = [1]
# 新增
a = 1, 2
a.append(3) #方法1
a.insert(2, 3) #方法2
# 刪除
a.pop() #方法1
a.pop(3)
del a[2:] #方法2
a.remove(3) #方法3
```

```python
## SET
# 建立
a = {1}
# 新增
a.add(2) #方法1
b = {2, 3}
a.update(b) #方法2
# 刪除
a.pop() #方法1
a.remove(1) #方法2
```

```python
## DICT
# 建立
a = {A: 1, B: 2}
b = dict(A=1, B=2)
# 取值
a['A']
# 新增(修改)
a['A'] = 0 #方法1
b = {C: 3, D: 4}
a.update(b) #方法2
# 刪除
a.pop('A') #方法1
del a['A'] #方法2
# 鍵
a.keys()
# 值
a.values()
# 項目
a.items()
```
