最後請印出此圓的半徑（Radius）、周長（Perimeter）和面積（Area）。
提示1：需import math模組，並使用math.pi。
提示2：輸出浮點數到小數點後第二位。

```python
import math

def calculate(radius):
    perimeter = 2 * math.pi * radius #計算圓的周長
    area = math.pi * radius ** 2 #計算圓的面積
    return perimeter, area #回傳周長和面積

radius = float(input('請輸入圓的半徑：'))

perimeter, area = calculate(radius)

print(f'半徑:{radius:.2f}')
print(f'周長:{perimeter:.2f}')
print(f'面積:{area:.2f}')
```

(1)請撰寫一程式，以不定數迴圈的方式讓使用者輸入西元年份，
然後判斷它是否為閏年（leap year）或平年。
其判斷規則如下：每四年一閏，每百年不閏，但每四百年也一閏。
(2)假設此不定數迴圈輸入-9999則會結束此迴圈。

```python
while True:
    year = int(input('請輸入西元年份:'))
    years = (year % 4 == 0 and year % 100 != 0) or year % 400 == 0
    if year == -9999:
        print('-9999')
        break
    elif years:
        print(year)
        print(f'{year} is a leap year.')
    else:
        print(year)
        print(f'{year} is not a leap year.')
```

輸入值如為2即表示針對2號候選人投票，如輸入其他值則視為廢票。
每次投完後需印出目前每位候選人的得票數，最後印出最高票者為當選人；
如最終計算有相同的最高票數者或無法選出最高票者，顯示【=> No one won the election.】。

```python
n = 0 #定義投票次數
n1 = 0 #定義得票數
n2 = 0
n3 = 0

while n != 5:
    num = input('請輸入候選人號碼:')
    n += 1
    if num == '1':
        n1 += 1
        print(f'No.1: Nami = {n1}')
        print(f'No.2: Chopper = {n2}')
        print(f'廢票 = {n3}')
    elif num == '2':
        n2 += 1
        print(f'No.1: Nami = {n1}')
        print(f'No.2: Chopper = {n2}')
        print(f'廢票 = {n3}')
    else:
        n3 += 1
        print(f'No.1: Nami = {n1}')
        print(f'No.2: Chopper = {n2}')
        print(f'廢票 = {n3}')

if n1 > n2:
    print('=>No.1 Nami won the election.')
elif n1 < n2:
    print('=>No.2 Chopper won the election.')
else:
    print('No one won the election.')
```

```python
n = int(input('請輸入一個值:'))

for i in range(1, n + 1):
    spaces = ' ' * (n - i)
    stars = '*' * (2 * i - 1)
    print(spaces + stars)
```
