```python
#製作一個function，搜尋list最大值
def fun(e):
    maxvalue = e[0]
    for i in range(len(e)):
        if e[i] > maxvalue:
            maxvalue = e[i]
    return maxvalue

e = [100, 3, 2]

fun(e)
```
