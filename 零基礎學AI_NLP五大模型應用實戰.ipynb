#Word2Vec

```python
import os, zipfile #解壓縮套件
if not os.path.exists("data"):
  f = zipfile.ZipFile("pttsmalldata.zip")
  f.extractall("./data")
  f.close()
```

```python
import glob, json
import pandas as pd
contents = []
for fn in glob.glob("data/pttsmalldata/*.json"):
  with open(fn, "r", encoding="utf-8") as f:
    data = json.load(f)
    contents.append(data["post_content"])
  df = pd.DataFrame({"contents": contents})
df.tail()
```

```python
df = df.head(10)
df
```

```python
import jieba #中文分詞套件
from urllib.request import urlretrieve
url = "https://github.com/fxsjy/jieba/raw/master/extra_dict/dict.txt.big"
urlretrieve(url, "dict.txt.big")
jieba.set_dictionary("dict.txt.big")
```

```python
import re
punct = set(u''':!),.:;?]}¢'"、。〉》」』】〕〗〞︰︱︳﹐､﹒﹔﹕﹖﹗﹚﹜﹞！），．：；？｜｝︴︶︸︺︼︾﹀﹂﹄﹏､～￠々‖•·ˇˉ―--′’”([{£¥'"‵〈《「『【〔〖（［｛￡￥〝︵︷︹︻︽︿﹁﹃﹙﹛﹝（｛“‘-—_…~/ －＊➜■─★☆=@<>◉é''')
#去掉全型標點符號
def process(content):
  content = re.sub(r'https?:\/\/.*[\r\n]*', '', content)
  content = " ".join(filter(lambda x: x not in punct, jieba.cut(content)))
  return content
content_cut = df["contents"].apply(process)
```

```python
corpus = [] #訓練語料
for c in content_cut:
  corpus += [s.lstrip().rstrip() for s in c.split("\n") if s.count(" ") >= 2]
corpus
```

```python
from keras.preprocessing.text import Tokenizer #語料計算套件
tok = Tokenizer()
tok.fit_on_texts(corpus)
tok.word_index
```

```python
corpus_seq = tok.texts_to_sequences(corpus)
V = len(tok.word_index) + 1 #語料量
dim = 100 #降維成多少特徵
window = 2 #取上文下文
```

```python
from keras.models import Sequential #模型支架
import keras.backend as K
from keras.layers import Embedding, Lambda, Dense
cbow = Sequential()
cbow.add(Embedding(V, dim, input_length=window*2))
cbow.add(Lambda(lambda x: K.mean(x, axis=1)))
cbow.add(Dense(V, activation="softmax"))
cbow.summary() #查看模型架構
```

```python
cbow.compile(loss="categorical_crossentropy", optimizer="adam") #建立模型
```

```python
from keras.preprocessing import sequence #序列套件
from keras.utils import to_categorical #One-hot encoding
def generate_data(corpus, window, V):
  for sentence in corpus:
    L = len(sentence)
    for i, word in enumerate(sentence):
      contexts = [] #上下文
      label = [] #機率
      start = i - window #開始
      end = i + window #結束
      c = [sentence[j] for j in range(start, end + 1) if not j == i and 0 <= j < L]
      contexts.append(c)
      label.append(word)
      x = sequence.pad_sequences(contexts, maxlen = window * 2)
      y = to_categorical(label, V)
      yield (x, y)
```

```python
g = generate_data(corpus_seq, window, V)
g.__next__()
```

```python
for epoch in range(10):
  loss = 0.
  for x, y in generate_data(corpus_seq, window, V):
    l = cbow.train_on_batch(x, y) #訓練模型
    loss += l
  print("epoch:", epoch)
  print("loss:", loss)
  print("*" * 50)
```

```python
f = open("word2vec.txt", "w", encoding="utf-8")
f.write("{} {}\n".format(V - 1, dim))
weights = cbow.get_weights()[0]
for word, i in tok.word_index.items():
  vec = " ".join(map(str, list(weights[i])))
  f.write("{} {}\n".format(word, vec))
f.close()
```
