```python
import pandas as pd
# 引入套件
df = pd.read_excel("dict_revised_2015_20240627/dict_revised_2015_20240627.xlsx")
# 讀取文字資料集
df = df[["字詞名", "釋義"]]
# 選擇需要欄位
```

```python
df.to_csv("data.csv", encoding="utf-8-sig", index=False)
# 儲存檔案
```

```python
import csv
import json
# 引入套件
def csv_to_json(csv_file, json_file):
  with open(csv_file, "r", encoding="utf-8-sig") as csvfile:
    csv_reader = csv.DictReader(csvfile)
    json_data = []
    # 建立空LIST
    for row in csv_reader:
      f = {
          "instruction": row["字詞名"],
          "input": "",
          "output": row["釋義"]
          }
      json_data.append(f)
  with open(json_file, "w", encoding="utf-8-sig") as jsonfile:
    jsonfile.write(json.dumps(json_data, indent=4, ensure_ascii=False))
    # ensure_ascii=False避免繁體中文編碼錯誤
csv_to_json("data.csv", "data.json")
# 調用函數
```
