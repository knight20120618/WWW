```python
import cv2 #匯入OpenCV模組

cap = cv2.VideoCapture(0) #讀取攝像鏡頭
#cap = cv2.VideoCapture('https://192.168.1.107:8080/video') #IP Webcam

if not cap.isOpened():
    print('CAMERA 404')
    exit()

while True:
    ret, frame = cap.read()
    if not ret:
        print('404')
        break
    frame = cv2.resize(frame, (480, 360)) #調整影像大小
    cv2.imshow('ZZZ', frame) #顯示影像資源

    if cv2.waitKey(1) == 27: #ESC按鍵
        break

cap.release() #釋放資源
cv2.destroyAllWindows() #關閉視窗
```
